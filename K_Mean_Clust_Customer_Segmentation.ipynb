```python
#Selecting Spending Score and Annual income for Clustering
clust_data = df.loc[:,["Annual Income (k$)" , "Spending Score (1-100)"]].values
clust_data
```

##K Mean Clustering

```python
from sklearn.cluster import KMeans
```

```python
#wwcs - Within Clusters Sum of Squares

wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(clust_data)

  wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11), wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Number of Clusters from above graph - 5 (as graph flasttens after 6)
```

```python
#Training K Mean Model
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0)

clusters = kmeans.fit_predict(clust_data)
clusters
```

```python
#5 Clusters - 0 1 2 3 4
```

```python
#Visualising Clusters using Scatter Plot
plt.scatter(clust_data[clusters==0,0], clust_data[clusters==0,1], s=40, c='blue', label='Group 1')
plt.scatter(clust_data[clusters==1,0], clust_data[clusters==1,1], s=40, c='orange', label='Group 2')
plt.scatter(clust_data[clusters==2,0], clust_data[clusters==2,1], s=40, c='red', label='Group 3')
plt.scatter(clust_data[clusters==3,0], clust_data[clusters==3,1], s=40, c='yellow', label='Group 4')
plt.scatter(clust_data[clusters==4,0], clust_data[clusters==4,1], s=40, c='magenta', label='Group 5')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=80, c='cyan', label='Centroids')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')  # legend outside the graph
plt.show()
```
